```python
import numpy as np
import pandas as pd
import requests, time
from requests.adapters import HTTPAdapter, Retry
import warnings

warnings.filterwarnings('ignore')
```

```python
API_KEY = '1234567890'
```

```python
session = requests.Session()
retries = Retry(total=5, backoff_factor=1, status_forcelist=[502, 503, 504])
session.mount('https://', HTTPAdapter(max_retries=retries))

def safe_get(url):
    try:
        response = session.get(url, timeout=10)
        response.raise_for_status()
        return response
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
        return None
```

```python
def get_info(content_type, pages=10):
    all_items = []
    for page in range(1, pages+1):
        url = f'https://api.themoviedb.org/3/discover/{content_type}?api_key={API_KEY}&page={page}'
        response = safe_get(url)
        if response.status_code != 200:
            print('got cooked')
            break
        
        data = response.json().get('results', [])
        
        for item in data:
            all_items.append({
                'id': item.get('id'),
                'title': item.get('title') or item.get('name'),
                'overview': item.get('overview'),
                'genres': item.get('genre_ids'),
                'popularity': item.get('popularity'),
                'vote_average': item.get('vote_average'),
                'release_date': item.get('release_date') or item.get('first_air_date'),
                'type': content_type
            })
        
        time.sleep(0.25)  #to respect rate limits
    return all_items
```

```python
def remove_alternate_spaces(s):
    if not isinstance(s, str):
        return s
    result = []
    space_count = 0
    for char in s:
        if char == " ":
            space_count += 1
            if space_count % 2 != 0:
                continue
        result.append(char)
    return ''.join(result)

only_crew = only_crew.apply(remove_alternate_spaces)
```
